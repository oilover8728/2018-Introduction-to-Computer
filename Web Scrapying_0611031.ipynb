```python
# 引入 Beautiful Socp 模組
from bs4 import BeautifulSoup
# 引入 requests 模組
import requests
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 透過 request 拿到html檔案
url = "https://www.youtube.com/?gl=TW"
res = requests.get(url)

# 透過 beautiful soup 來解析這個 html檔案
soup = BeautifulSoup(res.text)
number = input("你的Youtube首頁上最近觀看人數 前___ 的影片(約5上下)： ")
while number.isdigit() < 1:
    number = input("請輸入數字: ")
    # 必須輸入數字
```

```python
# 尋找標籤
import re # 多重分割

url_base = 'https://www.youtube.com/'
i = 0
j = 0
time= []
http = []
num = []
num2 = []
count = 0

data = soup.select('.yt-lockup-meta-info')
# for word in soup.select('.yt-lockup-video'):
#    word2 = word.select("a[rel='spf-prefetch']")
#    print(word2[0].get("title"))
for a in soup.select('a'):
#    print(a.title)
    i+=1
    if i > 21:
        if i < 256:
            count+=1
         # 除去youtube本身產生的網址
            if len(a['href']) == 20:
            # 避免讀取多餘的連結
                if (count%3) == 1:
                    j+=1
                    http.append(url_base + a['href'])
           #        print(url_base + a['href'])
for watch in data:
        time.append(watch.get_text().split("次")[0]+watch.get_text().split("次")[1])
        num.append(watch.get_text().split("次")[2])
        num2.append(re.split('[：,次 ]',watch.get_text())[2]+re.split('[：,次 ]',watch.get_text())[3]+re.split('[：,次 ]',watch.get_text())[4])
        # 以下為測試split有沒有錯誤
        
        # print(type(string))
        # a='729'
        # print(a.isdigit())
        # print(string)
       #  if string.isdigit() > 1:
       #      num2.append(re.split('[：,次]',watch.get_text())[2]+re.split('[：,次]',watch.get_text())[3]+re.split('[：,次]',watch.get_text())[4])
       #  elif string.isdigit() < 1:
       #      num2.append(re.split('[：,次]',watch.get_text())[2]+re.split('[：,次]',watch.get_text())[3])
# for k in range (0,j-10):
#     print(http[k])
#     print(time[k])
#     print(num2[k])
```

```python
temp=[]
temp2=[]
first=[]
index=[]

# 決定出前幾部觀看數高的影片
for s in range (0,int(number)):
    temp.append(0)
    temp2.append(0)
# print(number)
# print(temp)
# print(temp2)
for k in range (0,j-10):
    if int(num2[k]) > temp[0]:
        temp[0] = int(num2[k])
        temp2[0] = k

for s in range (1,int(number)):
    for k in range (0,j-10):
            if int(num2[k]) < temp[s-1]:
                if int(num2[k]) > temp[s]:
                    temp[s] = int(num2[k])
                    temp2[s] = k
# print(temp)
# print(temp2)

# 將得到的資料整理成長條圖要用的數字
for p in range (0,int(number)):
    index.append(p+1)
    first.append(num2[temp2[p]])
    print("第"+str(p+1)+"名") 
    print(http[temp2[p]])
    print(time[temp2[p]])
first.reverse()
index.reverse()
# print(index)

# 使用長條圖視覺化所得資料
matplotlib.rcParams['font.sans-serif'] = ['myriadpro-regular']

plt.figure(figsize=(int(number)*2.2,8))#圖形大小
plt.bar(index, first)
plt.xticks(index, fontname='Noto Sans CJK TC')
plt.show()
print("前"+number+"部觀看次數最高的影片與連結")
```
